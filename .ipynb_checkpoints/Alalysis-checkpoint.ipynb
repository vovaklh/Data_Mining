```python
import pandas as pd
import numpy as np
import sweetviz
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
data = pd.read_csv("Data/creditcard.csv")
data.head()
```

```python
# Get data shape
data.shape
```

**Exploratory Data Analysis**

>* Exploratory data analysis is an approach to analyzing data sets to summarize their main characteristics, often with visual methods. There are many packages in python to analyze,visualize,summerise the dataset. 
>* Here we are going to show you the automated EDA tool which do all analysis/visualization/summarization in a single shot!. It takes pandas dataframes and give result.

```python
report = sweetviz.analyze([data,'data'],target_feat='Class')
report.show_html('report.html')
```

**About sweetviz**

> Sweetviz is an open source Python library that generates beautiful, high-density visualizations to kickstart EDA (Exploratory Data Analysis) with a single line of code. Output is a fully self-contained HTML application.
